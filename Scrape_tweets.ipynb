```python
# setting up tweepy authorization 

import tweepy
import credentials as c

auth = tweepy.OAuthHandler(c.consumer_key, c.consumer_secret)
auth.set_access_token(c.access_token,c.access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)

text_query = '4more4nana'
count = 10000
tweets = tweepy.Cursor(api.search, q=text_query).items(count)

all_tweets = []
for tweet in tweets:
     all_tweets.append({'Date_created':tweet.created_at,'Tweet':tweet.text,'Source':tweet.source,'User_name':tweet.user.name})
all_tweets
```

```python
import pandas as pd
all_tweets_df = pd.DataFrame(all_tweets)
all_tweets_df = all_tweets_df.iloc[:, 0:3]
all_tweets_df.to_csv('4more4nana.csv', index=False)
all_tweets_df
```
